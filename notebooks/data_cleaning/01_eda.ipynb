```python
import os

import pandas as pd
import plotly.express as px
import seaborn as sns
import yaml
from dotenv import load_dotenv

from my_project.data.utils import columns_info_func

pd.options.display.max_columns = None
```

```python
load_dotenv(override=True)

DATA_DIR = os.getenv("DATA_DIR")
RESULTS_DIR = os.getenv("RESULTS_DIR")
CONFIGS_DIR = os.getenv("CONFIGS_DIR")

with open(os.path.join(CONFIGS_DIR, "data.yaml")) as f:
    data_config = yaml.safe_load(f)
```

```python
df = pd.read_csv(
    os.path.join(DATA_DIR, data_config["paths"]["raw_data_path"]), sep=";", decimal=","
)
df = df.set_index(data_config["columns"]["id"])
```

# Column statistics

```python
columns_info = columns_info_func(df)
```

```python
columns_info.query("is_categorical").sort_values("min_proportion").head(10)
```

## Missing values

```python
px.bar(
    columns_info[columns_info["missing_pct"] > 0],
    y="missing_pct",
    hover_data=["cumsum_pct"],
    labels={"index": "Columns", "value": "Proportion of missing values"},
    title="Proportion of missing values per column",
).show()
```

```python
df[columns_info.head(20).index.tolist()]
```

## Unique values in numeric columns

```python
numeric_columns_info = columns_info_func(
    df[df.select_dtypes(exclude=["object"]).columns],
    sort_by_column="n_unique",
    ascending=False,
)
```

```python
px.line(
    numeric_columns_info, y="n_unique", title="Unique values per numeric column"
).show()
```
